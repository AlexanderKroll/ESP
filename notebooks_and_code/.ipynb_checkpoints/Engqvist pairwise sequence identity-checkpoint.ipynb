```python
from Bio.Emboss.Applications import NeedleCommandline
import os
from os.path import join
import pandas as pd
import sys
import time
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

CURRENT_DIR = os.getcwd()

plt.style.use('CCB_plot_style_0v4.mplstyle');
c_styles      = mpl.rcParams['axes.prop_cycle'].by_key()['color']   # fetch the defined color styles
high_contrast = ['#004488', '#DDAA33', '#BB5566', '#000000']



def calculate_identity(arguments):
    [fasta_file_1, fasta_file_2, UID1, UID2] = arguments
    needle_cline = NeedleCommandline(asequence = fasta_file_1, bsequence = fasta_file_2,
                                     gapopen=10, gapextend=0.5,  filter = True)

    out = needle_cline()[0]
    out = out[out.find("Identity"):]
    out = out[:out.find("\n")]
    percent = out[out.find("(")+1 :out.find(")")-1]
    return(percent)
```

```python
identities = np.zeros((21,21))

engqvist_fasta_files = os.listdir(join(CURRENT_DIR, ".." ,"data", "enzyme_data", "Engqvist_UIDs"))


for j in range(len(engqvist_fasta_files)):
    UID1 = engqvist_fasta_files[j]
    fasta_file_1 = join(CURRENT_DIR, ".." ,"data", "enzyme_data", "Engqvist_UIDs", UID1)
    UID1 = UID1.split(".")[0]

    for i in range(len(engqvist_fasta_files)):
        UID2 = engqvist_fasta_files[i]
        fasta_file_2 = join(CURRENT_DIR, ".." ,"data", "enzyme_data", "Engqvist_UIDs", UID2)
        UID2 = UID2.split(".")[0]

        identities[j, i] = calculate_identity(arguments = [fasta_file_1, fasta_file_2, UID1, UID2])
```

```python
np.fill_diagonal(identities, 0)
```

```python
max_identity = [max(identities[i,:]) for i in range(len(engqvist_fasta_files))]
```

```python
max_identity
```

```python
plt.figure(figsize=[10,5])
plt.rcParams.update({'font.size': 28})
plt.hist(max_identity, bins = 30, rwidth=0.75, range = (0,100))
plt.title("Maximal pairwise sequence identities", size= 22)
plt.show()
```

```python
def pw_agreement(UID1, UID2):
    count = 0
    for sub in ["glycolate", "lactate", "2-hydroxyglutarate", "2-hydroxyoctanoate", "2-hydroxystearate", "mandelate"]:
        if list(df[sub].loc[df["protein"] == UID1])[0] == list(df[sub].loc[df["protein"] == UID2])[0]:
            count +=1
    return(count/6)
```
