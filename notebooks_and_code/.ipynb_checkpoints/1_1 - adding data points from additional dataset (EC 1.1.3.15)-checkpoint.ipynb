## 2. Creating data sets for model training with additional data:

### (a) Loading training and test sets:

```python
df_UID_MID_train = pd.read_pickle(join(CURRENT_DIR, ".." ,"data","enzyme_substrate_data", "df_UID_MID_train.pkl"))
df_UID_MID_test = pd.read_pickle(join(CURRENT_DIR, ".." ,"data", "enzyme_substrate_data", "df_UID_MID_test.pkl"))

df_UID_MID_train = df_UID_MID_train.sample(frac = 1, random_state = 1).reset_index(drop = True)
df_UID_MID_test = df_UID_MID_test.sample(frac = 1, random_state = 1).reset_index(drop = True)

df_UID_MID_train = df_UID_MID_train[df_UID_MID_train["evidence"] == "exp"]
df_UID_MID_test = df_UID_MID_test[df_UID_MID_test["evidence"] == "exp"]
```

### (b) Adding  data points from new set to training set:

Creating a DataFrame with all substrates:

```python
df_test_new_mols = pd.DataFrame(data= {"ECFP" : df_test_new["ECFP"], "molecule ID" : np.nan})

for ind in df_test_new_mols.index:
    if not pd.isnull(df_test_new["KEGG ID"][ind]):
        df_test_new_mols["molecule ID"][ind] =  df_test_new["KEGG ID"][ind]
    else:
        df_test_new_mols["molecule ID"][ind] =  df_test_new["InChiKey"][ind]

df_test_new_mols.drop_duplicates(inplace = True)
df_test_new_mols.reset_index(drop = True, inplace = True)
```

Adding data points to training set:

```python
print(len(df_test_new.loc[df_test_new["Binding"] == 1]), len(df_test_new.loc[df_test_new["Binding"] == 0]))

for ind in df_test_new.index:
    if not pd.isnull(df_test_new["KEGG ID"][ind]):
        df_UID_MID_train = df_UID_MID_train.append({"Uniprot ID" : df_test_new["Uniprot ID"][ind],
                             "molecule ID" :df_test_new["KEGG ID"][ind],  "type" : "engqvist",
                             "ECFP" : df_test_new["ECFP"][ind],
                                                    "Binding" : df_test_new["Binding"][ind]},
                                                   ignore_index = True)
    else:
        df_UID_MID_train = df_UID_MID_train.append({"Uniprot ID" : df_test_new["Uniprot ID"][ind],
                             "molecule ID" :df_test_new["InChiKey"][ind],  "type" : "engqvist",
                             "ECFP" : df_test_new["ECFP"][ind],
                                                    "Binding" : df_test_new["Binding"][ind]},
                                                   ignore_index = True)
            
df_UID_MID_train.loc[-120:]
```

Adding substrate representations to DataFrame:

```python
df_FPs = pd.read_pickle(join(CURRENT_DIR, ".." ,"data", "substrate_data", "df_ecfps.pkl"))

for ind in df_test_new_mols.index:
    df_FPs = df_FPs.append({"substrate ID": df_test_new_mols["molecule ID"][ind],
                            "ECFP" : df_test_new_mols["ECFP"][ind]}, ignore_index = True)
df_FPs
```

```python
df_UID_MID_train["ECFP"] = ""
df_UID_MID_test["ECFP"] = ""
for ind in df_FPs.index:
    sub_ID, ecfp = df_FPs["substrate ID"][ind], df_FPs["ECFP"][ind]
    df_UID_MID_train["ECFP"].loc[df_UID_MID_train["molecule ID"] == sub_ID] = ecfp
    df_UID_MID_test["ECFP"].loc[df_UID_MID_test["molecule ID"] == sub_ID] = ecfp
    
df_UID_MID_test = df_UID_MID_test.loc[df_UID_MID_test["ECFP"] != ""]
df_UID_MID_train = df_UID_MID_train.loc[df_UID_MID_train["ECFP"] != ""]
```

#### test set:

```python
df_UID_MID_test = drop_samples_without_mol_file(df = df_UID_MID_test)
df_metabolites_test, similarity_matrix_test = get_metabolites_and_similarities(df = df_UID_MID_test)
print(len(df_metabolites_test))

df_UID_MID_test["Binding"] = 1
df_UID_MID_test.reset_index(inplace = True, drop = True)

df_UID_MID_test = create_negative_samples(df = df_UID_MID_test, df_metabolites = df_metabolites_test,
                                          similarity_matrix = similarity_matrix_test)
df_UID_MID_test
```

Some of the newly added data points might have too many negative data points, since the dataset already contained negative data points. We are making sure that we will have a positive:negative ratio of 1:3 for these data points:

```python
new_enzymes = list(set(df_test_new["Uniprot ID"]))
```

```python
droplist = []
for enzyme in new_enzymes:
    mean_act = np.mean(df_UID_MID_train["Binding"].loc[df_UID_MID_train["Uniprot ID"] == enzyme])
    pos_samples = len(df_UID_MID_train.loc[df_UID_MID_train["Uniprot ID"] == enzyme].loc[df_UID_MID_train["Binding"] == 1])
    neg_samples = pos_samples * 2
    
    if mean_act < 0.33:
        help_df = df_UID_MID_train.loc[df_UID_MID_train["Uniprot ID"] == enzyme].loc[df_UID_MID_train["Binding"] == 0]
        droplist = droplist + list(help_df.index)[neg_samples:]
        
    print(enzyme, np.mean(df_UID_MID_train["Binding"].loc[df_UID_MID_train["Uniprot ID"] == enzyme]))
    
df_UID_MID_train.drop(droplist, inplace = True)

for enzyme in new_enzymes:
    print(enzyme, np.mean(df_UID_MID_train["Binding"].loc[df_UID_MID_train["Uniprot ID"] == enzyme]))
```

### (d) Mapping ECFPs and ESM-1b-vectors to different splits:

#### (d)(i) Mappings ECFPs:

```python
df_UID_MID_train.drop(columns = ["ECFP"], inplace = True)
df_UID_MID_train = df_UID_MID_train.merge(df_FPs, right_on = "substrate ID", left_on = "molecule ID", how = "left")

df_UID_MID_test.drop(columns = ["ECFP"], inplace = True)
df_UID_MID_test = df_UID_MID_test.merge(df_FPs, right_on = "substrate ID", left_on = "molecule ID", how = "left")
```

#### (d)(ii) Mappings ESM1b-vectors:

```python
df_train = df_UID_MID_train
df_test = df_UID_MID_test

UNIPROT_df = pd.read_pickle(join(CURRENT_DIR, ".." ,"data", "enzyme_data", "Uniprot_df_with_ESM1b.pkl"))

Uniprot_df = pd.DataFrame(data = {"Uniprot ID" : UNIPROT_df["Uniprot ID"],
                                 "ESM1b" : UNIPROT_df["ESM1b"]})
#adding new enzymes to Uniprot_df
for enzyme in new_enzymes:
    Uniprot_df = Uniprot_df.append({"Uniprot ID" : enzyme,
                                 "ESM1b" : list(df_test_new["ESM1b"].loc[df_test_new["Uniprot ID"] == enzyme])[0]},
                                  ignore_index=True)
    
df_train = df_train.merge(Uniprot_df, on = "Uniprot ID", how = "left")
df_test = df_test.merge(Uniprot_df, on = "Uniprot ID", how = "left")
```
